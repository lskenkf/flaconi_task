```python
import re
import os
import warnings
import datetime

import pandas as pd
import numpy as np
import pandasql as psql

from solution import *
from autokeras import StructuredDataRegressor
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None) # Let pandas print the entire dataframe
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
input_file= 'data\device_activations.csv'
data = pd.read_csv(input_file)
data['time'] = data['time'].apply(lambda dt: dt[:-5] + '00:00')
data = data.groupby(['device', 'time'])['device_activated'].agg('max').reset_index()
data['time'] = pd.to_datetime(data['time'])

new_data = data.set_index(
    ['time', 'device']
).unstack(
    fill_value=0
).asfreq(
    'H', fill_value=0
).stack().sort_index(level=1).reset_index()

#set(new_data.device_activated.values == data.device_activated.values)

data = extract_date_info(new_data)
# data = data.loc[data.week>26] we do not remove week 26 even if it has only 2 days records,
# because the input time could fall in this week.
```

```python
split_point = datetime.datetime(2016, 8, 23)
train_data = data[data.time <= split_point]
test_data = data[data.time > split_point]
```

```python
train_data = train_data.groupby(['device', 'day_name','hour']).agg(train_device_activated=('device_activated', 'max')).reset_index()
test_data = test_data.groupby(['device', 'day_name','hour']).agg(test_device_activated=('device_activated', 'max')).reset_index()
```

```python
join_data = train_data.merge(test_data,on=['device','day_name','hour'])
```

```python
from sklearn.metrics import classification_report
y_true = join_data.test_device_activated
y_pred = join_data.train_device_activated
print(classification_report(y_true, y_pred))
```
