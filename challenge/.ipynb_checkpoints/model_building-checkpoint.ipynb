```python
input_file= 'data\device_activations.csv'
data = pd.read_csv(input_file)
data['time'] = data['time'].apply(lambda dt: dt[:-5] + '00:00')
query = """
select 
    device, 
    time, 
    sum(device_activated) device_activated 
from data
    group by device,time
"""
data = psql.sqldf(query)
data['time'] = pd.to_datetime(data['time'])

new_data = data.set_index(
    ['time', 'device']
).unstack(
    fill_value=0
).asfreq(
    'H', fill_value=0
).stack().sort_index(level=1).reset_index()

#set(new_data.device_activated.values == data.device_activated.values)

data = extract_date_info(new_data)
# data = data.loc[data.week>26] we do not remove week 26 even if it has only 2 days records,
# because the input time could fall in this week.
```

```python
split_point = datetime.datetime(2016, 8, 23)
train_data = data[data.time <= split_point]
test_data = data[data.time > split_point]
```

```python
query = """
select 
    device , 
    week,
    day_name,
    sum(device_activated)  device_activated
from data 
where device is 'device_7'
group by device,week,day_name
"""

psql.sqldf(query)
```

```python
query = """
select 
    device , 

    sum(device_activated)  device_activated
from data 
group by device
order by device
"""

psql.sqldf(query)
```

```python
query = """
select 
    device train_device, 
    day_name train_day_name,
    hour train_hour,
    sum(device_activated)  train_device_activated
from train_data 
group by device,day_name,hour
order by device,day_name,hour
"""
train_data = psql.sqldf(query)
```

```python
query = """
select 
    device test_device, 
    day_name test_day_name,
    hour test_hour,
    sum(device_activated)  test_device_activated
from test_data 
group by device,day_name,hour
order by device,day_name,hour
"""
test_data = psql.sqldf(query)
```

```python
query = """
select test_device,test_day_name,test_hour,
case when test_device_activated = 0 THEN 0
else 1
end test_device_activated,

case when train_device_activated = 0 THEN 0
else 1
end train_device_activated


from test_data  te
left join train_data  tr
on te.test_device = tr.train_device 
and  te.test_day_name = tr.train_day_name 
and  te.test_hour = tr.train_hour 
"""
join_data = psql.sqldf(query)

from sklearn.metrics import classification_report
y_true = join_data.test_device_activated
y_pred = join_data.train_device_activated
print(classification_report(y_true, y_pred))
```
